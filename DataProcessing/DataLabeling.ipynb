```python
full_stop_len = len(df.iloc[(df[3].str[-1].values == ".") & (df[3].values != ".")])
comma_len = len(df.iloc[(df[3].str[-1].values == ",") & (df[3].values != ".")])
colon_len = len(df.iloc[(df[3].str[-1].values == ":") & (df[3].values != ".")])
```

```python
print("full_stop_len: ", full_stop_len)
print("comma_len: ", comma_len)
print("colon_len: ", colon_len)
```

```python
full_and_comma_len = 50000
colon_len = 9000
total_len = 500000
rest_len = total_len - colon_len - full_and_comma_len * 2
rest_len
```

```python
full_stop = df.iloc[(df[3].str[-1].values == ".") & (df[3].values != ".")]
full_stop["label"] = 1
print(full_stop.head())
full_stop[3] = full_stop[3].map(lambda x: x[:-1])
print(full_stop.head())
comma = df.iloc[(df[3].str[-1].values == ",")]
comma["label"] = 2
comma[3] = comma[3].map(lambda x: x[:-1])
colon = df.iloc[(df[3].str[-1].values == ":")]
colon["label"] = 3
colon[3] = colon[3].map(lambda x: x[:-1])
rest = df.iloc[(df[3].str[-1].values != ".") & (df[3].str[-1].values != ",") & (df[3].str[-1].values != ":")]
rest["label"] = 0
```

```python
data_frames = [full_stop[:full_and_comma_len], comma[:full_and_comma_len], colon[:colon_len], rest[:rest_len]]
df_merged = pd.concat(data_frames)
```
